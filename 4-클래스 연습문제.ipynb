### 전자계산기,가위바위보,친구관리

```python
#전자계산기#
class Calc:
    def plus(self,n1,n2) : return n1+n2
    def minus(self,n1,n2) : return n1-n2
    def times(self,n1,n2) : return n1*n2
    def div(self,n1,n2) : return n1/n2
    
class Input_hendle:
    def __init__(self):
        self.op_tuple=('+','-','*','/')
        
    def op_check(self,op):
        if op=='x': return -1
        elif op not in self.op_tuple: return 0
        else:  return 1
    
    def int_input(self):
        while(True):
            try :
                int_num=int(input("숫자를 입력하시오 : "))
                return int_num
            except ValueError:
                print("정수가 아닙니다!! 다시입력하세요!")
                
while(True) :
    Operator=input("연산자를 입력하시오 (x:종료) : ")
    if Input_hendle().op_check(Operator)==-1:
        break
    elif Input_hendle().op_check(Operator)==0:
        print("연산자를 잘못입력하셧습니다!!")
    else:
        num1=Input_hendle().int_input()
        num2=Input_hendle().int_input()
        if Operator=='+'  : result=Calc().plus(num1,num2)
        elif Operator=='-' : result=Calc().minus(num1,num2)
        elif Operator=='*' : result=Calc().times(num1,num2)
        elif Operator=='/' : result=Calc().div(num1,num2)
        print("결과 :",result)
```

```python
'''

뒤에 자성된 Volume. Tv, Radio, Speaker 클래스를 기반으로 확장

1.부모 클래스로 Power를 추가한다.
-powerON(), poweroff()

2.전원을 켰을때만 볼륨 조절을 할 수 있도록 한다.

3.볼륨조정은 MAX값을 100으로 하고 MIN 값을 0으로 한다.
'''
```

```python
class Volume:
    vol,volMAX,volMIN = 0,100,0
    def volumeUp(self, vol):
        self.vol += vol
        print("볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        
    def volumeDown(self, vol):
        self.vol -= vol
        print("볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
    
    def getMaxMin(self):
        return self.volMAX,self.volMIN

class Power:
    active = "OFF"
    def powerOn(self):
        print("전원 ON")
        self.active="ON"
        
    def powerOff(self):
        print("전원 OFF")
        self.active="OFF"

############################################

class TV(Volume,Power):
    def volumeUp(self, vol):
        self.vol += vol
        print("TV 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        
    def volumeDown(self, vol):
        self.vol -= vol
        print("TV 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다."% (vol, self.vol))

############################################

class Radio(Volume,Power):
    def volumeUp(self, vol):
        self.vol += vol
        print("Radio 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        
    def volumeDown(self, vol):
        self.vol -= vol
        print("Radio 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다."% (vol, self.vol))

############################################

class Speaker(Volume,Power):
    def volumeUp(self, vol):
        self.vol += vol
        print("Speaker 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        
    def volumeDown(self, vol):
        self.vol -= vol
        print("Speaker 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다."% (vol, self.vol))

#############################################

class UI:
    
    @staticmethod
    def drawLine():
        print("-"*10)
    
```
