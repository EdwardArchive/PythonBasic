## 클래스 작성 문법(대부분 함수에 대한 문법)

    1) self
        -현재 사용중인 객체의 주소를 저장하고 있는 참조 변수
        -read only
        
    2) 생성자(생성자 메소드)
        -메소드
        -객체를 생성하는 용도
        -객체를 처음만들때 단 한번만 호출됨
       
       소멸자(거의안씀)
        -마무리 작업
        -혹시나 잘못해서 주소를 잊어버려서 미쳐 못한 작업을 해줄수 있는 용도
    
    3) static(정적인 고정된)
        - 동일한 공간을 모든 객체들이 공동으로 사용
          
    4) _,__ (언더바)
        
        1) _ :
            - 기본 인터 프리터(명령 프롬프트)에서 사용
            - 값을 무시
            - 변수명과 함께 사용(접근권한) __ 이것도 함께 사용
    5) 상속
        - 너무 긴 코드 파악이 어렵거나 그런 코드들을 제사용 하기 위해 사용한다
        - 상속 설계시 고려할 점
            *부모 클래스는 최소한의 코드로 작성(모든 자식들이 공통적으로 가지고있는 기능으로만 설계가 되어야한다
            *중복된 코드가 있어서는 안된다
            *상속 받은 자식 클래스는 반드시 기능을 확장(추가)시켜줘야한다
        
        -Override(제정의), Overload(중복정의)
            #Override
                *부모한태 받은것을 부모를 바꾸지말고 자식을 바꾸는 방법이다
            #Overload(python 미지원)
                *똑같은 이름의 함수를 여러개 만들어서 쓰는것
                *편하긴 하지만 많은 리소스 비용이 든다
                
        -super(): 부모객체에 접근하기 위한 주소를 리턴
        
        -다중상속
        
        -추상 클래스(Abstarct Class) : 
            -기본적으로 추상 메소드를 한개 이상 가지고 있어야 한다
            -추상 메소드란 내용이 없는 메서드를 의미
            -추상 메소드는 자식 클래스가 강제로 오버라이딩을 해서 사용해야한다
            -추상 클래스는 객체 생성이 불가능 하다

```python
#static -객체안에 것들을 맴버라고 하는데 스테틱 메소드는 멤버 메소드라고 하지 않는다 
#       -객체 생성과 상관없이 프로그램이 실행되면 기본으로 메모리에 실행이 된다

class Info:
    nation='한국'
    name=str()
    age=int()
    height=float()
    telephone=int()
    
    def display(self):
        print("국적 : ",self.nation)
        print("이름 : ",self.name)
        print("나이 : ",self.age)
        print("신장 : ",self.height)
        print("휴대전화 : ",self.telephone)
        print("\n--------------------------")
    
    @staticmethod                                   #정식적으로 static method 가 되어버리기 때문에 self 를 받을수 없다
    def tell(ph):                                   #self 를 쓰지 않아도 static 메소드로 사용할수 있음
        #print("대표 전화번호 : {} 국가: {}".format(ph,Info.nation)) self 사용불가
        Info.telephone = ph
    
    @staticmethod
    def tel_show():
        print("대표 전화번호 : {} 국가: {}".format(Info.telephone,Info.nation)) 
```

```python
class Counter:
    n_cnt=0
    @staticmethod                                               #or self 로 사용하고 .n_cnt의 위치를 기본공간
    def __init__():
        Counter.n_cnt+=1 
    @staticmethod
    def printCall():
        print("객체의 수 : ",Counter.n_cnt)
    
    @classmethod
    def printCall2(c):
        print("{} called 객체의 수 : {}".format(c,c.n_cnt))
    
```

```python
class Calculator:
    def __init__(self,first=None,second=None):
        self.first=first
        self.second=second
    def plus(self):
        return self.first+self.second
    def minu(self):
        return self.first-self.second
    def mutiply(self):
        return self.first*self.second
    def div(self):
        return self.first/self.second
```

```python
class CalcV1(Calculator):
    def pow(self):
        return self.first**self.second
    
    def div2(self):
        if self.second==0:
            return "0으로 나눌수 없다"
        else:
            return self.first/self.second
#div method and div2 method 있기 때문에 아쉬움 낭비된다(메모리에 존재)
```

```python
calc1=Calculator(4,5)
print(calc1.plus())

calc2=CalcV1(30,4)
print(calc2.div())

calc2=CalcV1(30,0)
print(calc2.pow())
print(calc2.div2())
```

```python
#Override : 상속에서만 사용할수 있다
class CalcV1(Calculator):
    def pow(self):
        return self.first**self.second
    
    def div(self):                      #Mother class 와 same name method 로 바꿔주면된다
        if self.second==0:
            return "0으로 나눌수 없다"
        else:
            return self.first/self.second
```

```python
calc2=CalcV1(30,4)
print(calc2.div())

calc2=CalcV1(30,0)
print(calc2.pow())
print(calc2.div())
```

```python
class Person:
    say="난 사람이야"
    age="20"
    
    def __init__(self,age):
        print("Parson Class")
        self.age=age
    def info(self):
        print("나이 : {} 하고싶은말 : {}".format(self.age,self.say))

#########################################################

class Employee(Person):
    say="일하는 동물"
    subject="근로자"
    
    def __init__(self):
        print("Employee Class")
        
    def einfo(self):
        print(self.age)
        print(self.say)
        print(super().age)
        print(super().say)

#########################################################
```

```python
class Person:
    say="난 사람이야"
    __age="20"
    
    def __init__(self,age):
        print("Parson Class")
        self.__age=age
        
    def info(self):
        print("나이 : {} 하고싶은말 : {}".format(self.__age,self.say))
    
    def getter(self):
        return self.__age
    
    def setter(self,age):
        #검증이 가능하다 x=인수 라고 하면 검증할 시간이 없거나 검증코드를 반복하여 사용해야 하기 때문에 메소드를 사용하는것이 좋다
        self.__age=age

#########################################################

class Employee(Person):
    say="일하는 동물"
    subject="근로자"
    
    def __init__(self):
        print("Employee Class")
        
    def einfo(self):
        print(self.getter())
        print(self.say)

```

```python
class Volume:
    vol = 0
    def volumeUp(self, vol):
        self.vol += vol
        print("볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        
    def volumeDown(self, vol):
        self.vol -= vol
        print("볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        
############################################

class TV(Volume):
    def volumeUp(self, vol):
        self.vol += vol
        print("TV 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        
    def volumeDown(self, vol):
        self.vol -= vol
        print("TV 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다."% (vol, self.vol))

############################################

class Radio(Volume):
    def volumeUp(self, vol):
        self.vol += vol
        print("Radio 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        
    def volumeDown(self, vol):
        self.vol -= vol
        print("Radio 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다."% (vol, self.vol))

############################################

class Speaker(Volume):
    def volumeUp(self, vol):
        self.vol += vol
        print("Speaker 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        
    def volumeDown(self, vol):
        self.vol -= vol
        print("Speaker 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다."% (vol, self.vol))

#############################################

tv=TV()
radio=Radio()
spk=Speaker()

tv.volumeUp(18)
tv.volumeDown(7)

radio.volumeUp(22)
radio.volumeDown(18)

spk.volumeUp(10)
spk.volumeDown(7)
```

```python
#추상 클래스
import abc

class Volume:
    vol=0
    
    @abc.abstractmethod
    def volumeup(self,vol):
        pass
    
    def volumedown(self,vol):
        pass
    

class TestVolume(Volume):
    def volumeup(self,vol):
        self.vol=vol
        print("소리 증가",self.vol)
    
    def volumedown(self,vol):
        self.vol=vol
        print("소리 감소",self.vol)
##################################################

t1=Volume()
t1.volumeup(100)

t2=TestVolume()
t2.volumeup(100)
```

## 클래스의 관계

1) is a 관계 : 상속관계, 일반화 관계
    -특정 클래스 객체가 다른 클래스 객체에게 자신의 멤버를 물려주는 상속관계
    UML에서는 비어있는 화살표로 표현 (밑 자식 위에는 부모) ^
    강한 상속 관계
    
2) has a 관계 : 연관관계, 포함관계
    -특정 클래스 네에서 해당 클래스의 맴버처럼 다른 클래스 객체를 포함하고 있는 경우
    
    -Aggregation : 약한관계
    -Composition : 강한관계

3) UML:

```python
#has a 관계
class Handle:
    qent=0
    
    def leftTurn(self,q):
        self.qent=q
        return "좌회전"
    
    def rightTurn(self,q):
        self.qent=q
        return "좌회전"
    
    def straight(self,q):
        self.qent=q
        return "좌회전"
################################

class Car:
    turnShow="정지"
    
    def __init__(self,ownerName):
        self.ownerName=ownerName
        self.handle=Handle()
    
    def turnhandle(self,q):
        if q>0:
            self.turnShow = self.handle.rightTurn(q)
        elif q<0:
            self.turnShow = self.handle.leftTurn(q)
        elif q==0:
            self.turnShow = self.handle.straight(q)
##############################################
txs=Car('tom')
txs.turnhandle(100)
print(txs.ownerName + "의 회전량은"+ txs.turnShow + str(txs.handle.qent))
```

```python
#약한관계

class Fridge:
    isOpen = False
    foods = []
    
    def open(self):
        self.isOpen = True
        print("냉장고 문 열기")
    
    def close(self):
        self.isOpen = False
        print("냉장고 문 닫기")
    
    def put(self, thing):
        if self.isOpen:
            self.foods.append(thing)
            print("냉장고에 음식이 들어감")
            self.list()
        else:
            print("냉장고 문이 닫혀있어서 음식을 넣을 수 없음")
    
    def list(self):
        for f in self.foods:
            print("-", f.irum, f.ex_data)
        print()
    
class FoodData:
    def __init__(self, irum, ex_data):
        self.irum = irum
        self.ex_data = ex_data
        
##############################################

f = Fridge()
apple = FoodData("사과", "2019-12-16")

f.open()
f.put(apple)
f.close()
```
